```python
# Arange
print("ARANGE\n")
a = np.arange(10)       # Funciona igual que [start : end : step]
print(a,"Dim:",a.ndim)
a = np.arange(5,16)
print(a)
a = np.arange(0,30,3)
print(a)

print("\nRANDOM PERMUTATION\n")
# Random permutation
a = np.random.permutation(np.arange(10)) # Desordena el arange
print(a)
print(np.random.randint(0,100,10)) # RandomInt entre [a,b) y c numeros 
a = np.random.rand(10)             # Genera n numeros entre 0 y 1
b = np.random.rand(2,4,2,2)        # Para arreglos multidimensionales random
print(a)
a = np.random.randn(10000)         # Distribucion gaussiana
plt.hist(a,bins=100,color='k')

print("\nRESHAPE\n")
# Reshape
a = np.arange(20).reshape(2,2,5)    # Reordena los datos en las dimensiones requeridas
print(a)
print("ZEROS:",np.zeros((2,2)))     # Regresa un arreglo de 0's con la shape dada
print("ONES:",np.ones((2,2)))       # Regresa un arreglo de 0's con la shape dada
```

```python
# SLICING

# normal
print("normal")
a = [0,1,2,3]
b = a[:]                 # Se crea una copia de la lista
print("B:",b,"A:",a)
b[0] = -10
print("B:",b,"A:",a)     # Al modificarlo, solo modifica b

# numpy
print("\nnumpy")
a = np.arange(4)
b = a[:]                 # Se crea un "puntero" al slicing de la lista
print("B:",b,"A:",a)
b[0] = -10
print("B:",b,"A:",a)     # Al modificar b, tambien modifica a
c = a[:].copy()          # Para crear copia se usa copy()
c[0] = 0
print("A:",a,"C:",c)

print()
a = np.arange(20)
print(a[::3])
print(a[::-1])           # Invertir el arreglo


a = np.arange(10)        # Coso raro para sacar indices
b = (a > 4) * np.ones(a.size)
print(b)

a = np.random.randint(0,10,20)
print(a)
indices = np.argwhere(a < 3)   # Argwhere te dice en que indices se cumple la condicion requerida
print("Cuantos:",indices.size)
indices = indices.reshape(indices.size)    # Reshape para guardarlo en un unico array
print("Indices:",indices,)

a = np.round(10*np.random.rand(3,4))       # Funciona basicamente igual que randint
```

```python
# SLICING IN NUMPY CHIDO

print(a)
print("Fila 2, Columna 1 :",a[2,1])   # Fila 2, Columna 1 : 3
print("Columna 2:",a[:,2])
print("Fila 0:",a[0,:])
print("Fila 0 a 1, Columna 1 a 3\n",a[0:2,1:4])

# Sort
print("\n",a)
a.sort(axis = 1)            # Axis tomade referencia la i-esima dimension,
print("Sort by rows\n",a)   # Axis 1 serian filas, 0 serian columnas
a.sort(axis = 0)            # Al aumentar dimensiones, puede ordenarse
print("Sort by cols\n",a)   # respecto a la dimension que queramos
```

```python
# Broadcast

a = np.random.randint(0,10,6).reshape(2,3)
print("A\n",a,"\n")
print(a + 3)                    # Sumar a todos los elementos
print(a * 2)                    # Multiplicar a todos los elementos
b = np.random.randint(0,10,6).reshape(2,3)
print("\nB\n",b,"\n")
print(a+b)                      # Suma elemento por elemento correspondiente
print()

# Concatenacion de matrices
c = np.hstack((a,b))              # Las acomoda horizontalmente
print(c)  
d = np.vstack((a,b))              # Las acomoda verticalmente
print(d)
```

```python
# COPIAS

a = np.arange(20)
print(a)
b = a[[2,5,6]]         # Lista con indices
print(b)
b[0] = -1
print(a,b)             # Se crea una copia, no se modifica el original

c = a[(a>5)&(a<14)]    # Indices que cumplan la(s) condicion(es)
print(c)
c[0] = -1
print(a,c)             # Se crea una copia, no se modifica el original
```

$ NUMPY \ \ -\ \ LINEAL\ ALGEBRA $

```python
import numpy.linalg as la

a = np.random.randint(1,4,6).reshape(2,3)
b = np.random.randint(1,4,9).reshape(3,3)

print("Matriz A\n",a)
print("\nMatriz B\n",b)
print("Determinante B\n",la.det(b))   # Determinante
print("INVERSA B\n",la.inv(b))        # Inversa de la matriz
print("Transpuesta A\n",a.T)          # Transpuesta de la matriz

print(" A X B\n",a@b)                 # Multiplicacion matricial
```

$ F(X,Y) = \sin(\frac{1}{2}x^2\ -\ \frac{1}{4}y^2\ +\ 3)\cos(2x\ +\ 1\ -\ e^y)$

```python
func = lambda xy: np.sin((1/2)*(xy[0]**2) - (1/4)*(xy[1]**2) + 3)*np.cos(2*xy[0] + 1 - np.e**xy[1])
n = 100
X = np.linspace(-2,2,n)
Y = np.linspace(-2,2,n)

Z = np.zeros((n,n))

for ix,x in enumerate(X):
    for iy,y in enumerate(Y):
        Z[ix,iy] = func([x,y])
        
plt.contour(X,Y,Z,50)            # Visualización en 3D
```

```python
plt.contourf(X,Y,Z,15)            # Visualización en 3D solida
```
